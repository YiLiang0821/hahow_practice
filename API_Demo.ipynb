```python
import googlemaps
from datetime import datetime
```

```python
gmaps = googlemaps.Client(key = 'AIzaSyDo0eUiEK-WIdUKXr3h58CWp-JHzQleDdw')
```

```python
cities = ['臺北市','新竹市','臺中市','高雄市','花蓮縣'] #放你想要的縣市，搜尋該縣市中的標的物
```

```python
geocode_result = gmaps.geocode('新竹市')  #轉成經緯度
geocode_result                           #要location的資訊
```

```python
loc = geocode_result[0]['geometry']['location']  #取得經緯度
loc
```

```python
gmaps.places_radar(keyword = '茶湯會',location = loc, radius = 25000)  #撈出以新竹市為中心，半徑25000公尺(預設為公尺)的茶湯會
                                                                      #得到id裡面的資料
```

```python
alist = []      #存放店家id(palce_id)
for i in cities:
    geocode_result = gmaps.geocode(i)
    loc = geocode_result[0]['geometry']['location']
    print(i +'半徑25000m的茶湯會數量:'+str(len(gmaps.places_radar(keyword = '茶湯會',location = loc, radius = 25000)['results'])))
    
    for palce in gmaps.places_radar(keyword = '茶湯會',location = loc, radius = 25000)['results']:
        alist.append(palce['place_id'])

alist = list(set(alist))                  #可能有重複的，用set篩選
print('總共店家數量：'+str(len(alist)))
```

```python
gmaps.place(place_id = alist[0],language='zh-TW')['result']  #試著印出店家資訊
```

```python
stories_info = []

for id in alist:
    stories_info.append(gmaps.place(place_id = id,language='zh-TW')['result'])
```

```python
output = pd.DataFrame.from_dict(stories_info)
output.head(5)
```

```python
output['geometry'][0]
```

```python
output['lat'] = output['geometry'].map(lambda x : x['location']['lat'])    #把經緯度拉出來
output['lng'] = output['geometry'].map(lambda x : x['location']['lng'])
output2 = output.drop(['geometry'],axis=1) #del output['geometry'] is the same
```
